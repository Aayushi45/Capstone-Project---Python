<b>Task 1: Make three dataframe in python and save them as three .csv files</b>

```python
import pandas as pd
#Creating Project Dataframe
# Creating a dictionary with given data
proj_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],
    'Status' : ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']
}

# Creating Project DataFrame from the dictionary
proj_df = pd.DataFrame(proj_data)

# Save the DataFrame to a CSV file
proj_df.to_csv('Project Dataframe.csv', index=False)

#Creating Employee Dataframe
# Creating a dictionary with given data
emp_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxemberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}

# Creating Employee DataFrame from the dictionary
emp_df = pd.DataFrame(emp_data)

# Save the DataFrame to a CSV file
emp_df.to_csv('Employee Dataframe.csv', index=False)

#Creating Seniority Level Dataframe
# Creating a dictionary with given data
sl_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level': [2, 2, 3, 2, 3]
}

# Creating Seniority level DataFrame from the dictionary
sl_df = pd.DataFrame(sl_data)

# Save the DataFrame to a CSV file
sl_df.to_csv('Senrity Level Dataframe.csv', index=False)

display(proj_df)
display(emp_df)
display(sl_df)
```

<b>Task 2: The cost column in the dataframe “Project” has some missing values. Compute these missing values. Replace the missing values by running average. Use the “For” loop for this task.</b>

```python
#Loading Project Dataframe
proj_df = pd.read_csv('Project Dataframe.csv')

#Creating a list of running averages using 'Cost' column in the project dataframe
running = 0
count = 0

# Calculating the average cost
average_cost = round(proj_df['Cost'].mean(), 2)

# Replacing the missing values with average cost using for loop
for i in range(len(proj_df)):
    #isnull() function checks if there are any null values
    if pd.isnull(proj_df.loc[i, 'Cost']): 
        proj_df.loc[i, 'Cost'] = average_cost

    #else function iterates the loop if there are no missing values
    else:
        running += proj_df.loc[i, 'Cost']
        count += 1

display("Updated project dataframe:",proj_df)
```

<b>Task 3: Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column.</b>

```python
#Loading Project Dataframe
emp_df = pd.read_csv('Employee Dataframe.csv')

#.split() function splits the name
empl_df = emp_df["Name"].str.split()
print("Split Names:\n",empl_df,"\n")

#Splitting the "Name" column into two different columns
emp_df[["First Name", "Last Name"]] = emp_df["Name"].str.split(' ', expand=True)

#Dropping the "Name" column drop()
emp_df.drop("Name", axis = 1, inplace=True)

# Reordering columns
cols = list(emp_df.columns)

#inserting the column in 1st index...
cols.insert(1, cols.pop(4))
cols.insert(2, cols.pop(5))
#used to reorder the columns of a DataFrame.
emp_df = emp_df.reindex(columns=cols)

display("Updated Employee DataFrame: ", emp_df)
```

<b>Task 4: Join all three dataframes in one single dataframe. Name it “Final”</b>

```python
#Using .merge() function to merge all the dataframes.
Final = emp_df.merge(proj_df,on='ID').merge(sl_df,on='ID')
Final.to_csv('Final.csv', index= False)
display(Final)
```

<b>Task 5: Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects.</b>

```python
#Loading Final Dataframe
final_df = pd.read_csv('Final.csv')

#Adding bonus column 
final_df['Bonus'] = final_df.apply(lambda row: row['Cost'] * 0.05 if row['Status'] == 'Finished' else 0, axis=1)

final_df.to_csv('final.csv', index=False)
display(final_df)
```

<b>Task 6: Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.</b>

```python
# Demoting fail project status employees 
final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] -= 1

# Deleting the employee records having designation level > 4
to_delete = 0
for index, row in final_df.iterrows():
    if row["Designation Level"] > 4: #Checking if any of the employees have designation level > 4
        to_delete.append(index)      #Adding the index at the end of the list `to_delete` if designation level > 4
        final_df.drop(to_delete, inplace=True)  #Removing the index from the Dataframe

# Saving the updated DataFrame in csv
final_df.to_csv('final.csv', index=False)

display("Updated Final DataFrame: ",final_df)
```

<b>Task 7: Add “Mr.” and “Mrs.” to the first name column and drop the gender column.</b>

```python
#Adding "Mr." to the first name of males and "Mrs." to the first name of females
final_df['First Name'] = final_df['Gender'].apply(lambda x: 'Mr. ' if x == 'M' else 'Mrs. ') + final_df['First Name']

#Dropping the Gender column from the dataframe
final_df = final_df.drop("Gender", axis=1)

display("Updated Final DataFrame: ",final_df)
```

<b>Task 9: Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost</b>

```python
#Grouping by employee ID and first name, and calculating the total project cost by summing up the cost column
TotalProjCost = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

#Renaming the columns
TotalProjCost.columns = ['ID', 'First Name', 'Total Cost']

#Converting the TotalProjCost Dataframe  to a .csv file
TotalProjCost.to_csv("Total_Project_Cost_Dataframe.csv")

display("TotalProjCost DataFrame: ",TotalProjCost)
```
